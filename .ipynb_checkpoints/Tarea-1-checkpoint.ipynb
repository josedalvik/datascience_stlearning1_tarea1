```python
#Importando librerías
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import tensorflow as tf
```

```python
#●	Crear una función para entrenar un modelo de regresión lineal de una variable y = mx +b.
def regresion_lineal(x, y, epochs, imprimir_error_cada, lr, mb):
    
    #○	Crear una matriz de 2 columnas ,la primera columna corresponde al vector de datos “x” y la segunda columna de la matriz para todas las filas es igual a 1
    matriz_auxiliar = np.hstack((x, np.ones(x.shape) ))
    
    #○	Inicializar los parámetros del modelo en un vector (m,b), esto es equivalente a empezar el proceso con una recta inicial la cual en cada iteración actualizaremos hasta encontrar una que aproxime de buena manera los datos x,y
    mb = np.array(mb).reshape(2,1)
    
    #Vector de errores
    errores = np.zeros(epochs)
    
    #Estructura de datos
    estructura_datos = {}
    
    #Iteraciones, de 1 a epochs
    for iteracion in range(1, epochs+1):
        
        #■	Calcular yhat(predicción o estimación) para todas las observaciones de manera simultánea(vectorizada)  utilizando el modelo correspondiente a la iteración(es decir , los valores de m,b): esto produce un vector yhat con el mismo número de elementos que y.
        yhat = np.matmul(matriz_auxiliar, mb)
        
        #■	Calcular el error o costo usando: y, y_hat
        error = np.nanmean((y - yhat)**2)/2
        
        if iteracion%imprimir_error_cada == 0:
            print("error (", iteracion,"):", error)

        #■	Almacenar en un vector el error de cada iteración. 
        errores[iteracion-1] = error
        
        #■	Calcular el gradiente del error respecto de cada parámetro con las expresiones
        gradiente = np.nanmean((yhat - y)*matriz_auxiliar, axis=0).reshape(2,1)
        
        #■	Actualizar los parámetros del modelo con la expresión
        mb = mb - (lr*gradiente)
        
        estructura_datos[iteracion] = {'yhat': yhat, 'gradiente': gradiente, 'mb': mb}
        
        #No continua evaluando si el error es infinito
        if np.isinf(error):
            break
    
    return (errores, estructura_datos)
        
print("\nRegresión lineal con el primer valor más predictivo")
epochs = 25000
imprimir_error_cada = 5000
lr = 0.001
y = my_reshape(data_training[:,0])
x = my_reshape(data_training[:,columna_1st_predictivo])
resultados_1st_predictivo = regresion_lineal(x, y, epochs, imprimir_error_cada, lr, [0.001, 0.001])

print("\nRegresión lineal con el segundo valor más predictivo")
epochs = 20
imprimir_error_cada = 4
lr = 0.000001
y = my_reshape(data_training[:,0])
x = my_reshape(data_training[:,columna_2nd_predictivo])
resultados_2nd_predictivo = regresion_lineal(x, y, epochs, imprimir_error_cada, lr, [200, 50000])

```

```python
class ModeloLineal:
    def __init__(self, x, y, tamaño_muestra, lr, epochs, batch_size):
        self.errores = []
        self.mb = tf.Variable([[0.0], [0.0]], dtype=tf.double, name="pendiente_intercepto")
        self.x_aux = np.hstack(( my_reshape(x), np.ones((x.size,1)) ))
        self.tamaño_muestra = tamaño_muestra
        self.lr = lr
        self.epochs = epochs
        self.batch_size = batch_size
        self.total_iteraciones = int(tamaño_muestra/batch_size)
        self.y = y
    
    def costo(self, y_real,y_aprox):
      """Funcion de costo/perdida/error
        para este caso: mean squared error(MSE)
      """
      return 1/2*tf.reduce_mean(tf.math.square(y_real - y_aprox) )
    
    def predecir(self):
        for epoch in range(self.epochs):
          #shuffle(x,y) # buena practica (ordenar aleatoriamente)

            for i in range(self.total_iteraciones):  #muchas iteraciones por epoch 
              muestra_inicio = i*self.batch_size
              muestra_fin = muestra_inicio + self.batch_size
              x_mb =  np.array(self.x_aux[muestra_inicio:muestra_fin,:]).reshape([-1,2])
              y_mb = np.array(self.y[muestra_inicio:muestra_fin]).reshape([-1,1])
                
              with tf.GradientTape() as grad_tape:
                grad_tape.watch(self.mb)
                error = self.costo(y_mb,  tf.matmul(x_mb, self.mb))

              # calcular el gradiente de la funcion de costo respecto de los parametros
              grad_mb = grad_tape.gradient(error, self.mb)

              # actualizar los parametros dando un paso en direccion contraria al gradiente
              self.mb.assign(self.mb - self.lr*grad_mb)

              # almacenar o procesar cualquier informacion relevante
              self.errores.append(error.numpy())

              print("Iteracion {}, error:{}".format(i,error))
              print("    modelo:y={}x+{}".format(self.mb[0,0].numpy(),self.mb[1,0].numpy()))

tamaño_muestra = 20
lr = 0.001
epochs = 30
batch_size = 10
y1 = my_reshape(data_training[:,0])
x1 = my_reshape(data_training[:,columna_1st_predictivo])

modelo_variable1 = ModeloLineal(x1, y1, tamaño_muestra, lr, epochs, batch_size)
modelo_variable1.predecir()
```

```python
#●	Crear una función que nos permita visualizar con matplotlib cómo cambia el error en el tiempo: crear una función que tome como parámetro el vector de errores generados por la función de entrenamiento y grafique en el eje “x” el número de iteración y en el eje “y”  el error para esa iteración:

def graficar_error(errores, eje_y_titulo):
    t = np.arange(1,errores.shape[0]+1,1)
    errores = np.array(errores)
    f, ax = plt.subplots(1)
    plt.title('Error por iteración')
    plt.xlabel('Iteración')
    plt.ylabel(eje_y_titulo)
    ax.plot(t, errores)
    ax.set_ylim(ymin=0)
    plt.show(f)
    
print("\nPrimer valor más predictivo")
graficar_error(resultados_1st_predictivo[0], columnas_nombre[columna_1st_predictivo])

print("\nSegundo valor más predictivo")
graficar_error(resultados_2nd_predictivo[0], columnas_nombre[columna_2nd_predictivo])
```

```python
#●	Crear una función que nos permita visualizar con matplotlib cómo evoluciona el modelo entrenado en el tiempo 

def graficar_modelo(n, resultados, columna_predictiva):
    
    for index in np.arange(n, len(resultados), n):
        mb = resultados[index]['mb']
        
        plt.scatter(data_training[:,columna_predictiva], data_training[:,0], alpha=0.5)
        plt.plot(data_training[:,columna_predictiva], data_training[:,columna_predictiva]*mb[0] + mb[1], linestyle='solid', color="g")
        
        plt.xlabel(columnas_nombre[columna_predictiva])
        plt.ylabel(columnas_nombre[0])
        plt.title("Iteración "+str(index))
        plt.show()

print("\nEvolución de modelo del primer valor predictivo\n")
graficar_modelo(5000, resultados_1st_predictivo[1], columna_1st_predictivo)
print("\nEvolución de modelo del segundo valor predictivo\n")
graficar_modelo(4, resultados_2nd_predictivo[1], columna_2nd_predictivo)
```

```python
#●	Para cada una de las variables x seleccionadas, usar scikit-learn para entrenar un modelo de regresión lineal

def scikitlearn_regresion_lineal(columna_predictiva):
    X = my_reshape(data_training[:,columna_predictiva])
    y =  my_reshape(data_training[:,0])
    scikitlearn_modelo = LinearRegression().fit(X, y)
    print("score:", scikitlearn_modelo.score(X, y))
    print("coef:", scikitlearn_modelo.coef_)
    print("intercept:", scikitlearn_modelo.intercept_)
    return scikitlearn_modelo

print("Regresión lineal con scikit-learn")

#Primer variable con más valor predictivo
print("\nPrimer variable con más valor predictivo\n")
scikitlearn_modelo_1ra_variable = scikitlearn_regresion_lineal(columna_1st_predictivo)

#Segunda variable con más valor predictivo
print("\nSegunda variable con más valor predictivo\n")
scikitlearn_modelo_2nd_variable = scikitlearn_regresion_lineal(columna_2nd_predictivo)
```

```python
#●	Para cada variable independiente x elegida, crear una función que usando el modelo entrenado manualmente y el modelo de scikit-learn(enviados como parámetros a la funcion) calcule la estimación o predicción de cada uno de estos y devuelva como resultado una predicción estimada promediando las predicciones de los 2  modelos, la función debe recibir como parámetro adicional un vector de cualquier tamaño de x  y devolver 3 vectores cada uno del mismo tamaño del vector x.
precio_real = my_reshape(data_validation[:,0])

def obtener_modelo_manual(resultados):
    index = len(resultados)
    return resultados[index]['mb']

def predecir_modelo_regresion_lineal(x, mb):
    matriz_auxiliar = np.hstack((x, np.ones(x.shape) ))
    r = np.matmul(matriz_auxiliar, mb)
    return r

def calcular_estimacion(scikitlearn_modelo_variable, modelo_manual, x):
    scikitlearn_predict = scikitlearn_modelo_variable.predict(x)
    modelo_manual_predict = predecir_modelo_regresion_lineal(x, modelo_manual)
    promedio = np.nanmean(np.array([scikitlearn_predict, modelo_manual_predict]), axis=0)
    return (scikitlearn_predict,modelo_manual_predict, promedio)

x_1ra_variable = my_reshape(data_validation[:,columna_1st_predictivo])
x_2nd_variable = my_reshape(data_validation[:,columna_2nd_predictivo])
estimacion_1ra_variable = calcular_estimacion(scikitlearn_modelo_1ra_variable, obtener_modelo_manual(resultados_1st_predictivo[1]), x_1ra_variable) 
estimacion_2nd_variable = calcular_estimacion(scikitlearn_modelo_2nd_variable, obtener_modelo_manual(resultados_2nd_predictivo[1]), x_2nd_variable) 
      
```

```python
# ●	Usando el 20% de los datos de validación, aplicamos los modelos predictivos para estimar la variable dependiente  y usando los valores reales de “y” calculamos el error de cada modelo.

def grafica_barras(scikitlearn, modelo_manual, promedio, titulo):
    plt.rcParams["figure.figsize"] = (4, 3)
    fig = plt.figure()
    ax = fig.add_axes([0,0,1,1])
    titulos = ['scikitlearn', 'modelo_manual', 'promedio']
    error = [np.nanmean(scikitlearn), np.nanmean(modelo_manual), np.nanmean(promedio)]
    ax.bar(titulos,error)
    plt.title(titulo)
    plt.show()
    
precio_real = my_reshape(data_validation[:,0])

#Errores para la primera variable con el valor más predictivo
error_1ra_variable_scikitlearn = precio_real - estimacion_1ra_variable[0]
error_1ra_variable_modelo_manual = precio_real - estimacion_1ra_variable[1]
error_1ra_variable_promedio = precio_real - estimacion_1ra_variable[2]
grafica_barras(error_1ra_variable_scikitlearn, error_1ra_variable_modelo_manual, error_1ra_variable_promedio, "Error para la primera variable")

#Errores para la segunda variable con el valor más predictivo
error_2nd_variable_scikitlearn = np.sqrt((precio_real - estimacion_2nd_variable[0])**2)
error_2nd_variable_modelo_manual = np.sqrt((precio_real - estimacion_2nd_variable[1])**2)
error_2nd_variable_promedio = np.sqrt((precio_real - estimacion_2nd_variable[2])**2)
grafica_barras(error_2nd_variable_scikitlearn, error_2nd_variable_modelo_manual, error_2nd_variable_promedio, "Error para la segunda variable")
```

```python
#Comparación con gráfica de líneas

def grafica_lineas(precio, scikitlearn, modelo_manual, columna_predictiva, titulo):
    index = np.arange(0, precio.shape[0], 1)
    
    plt.rcParams["figure.figsize"] = (12, 9)
    f, ax = plt.subplots(1)
    plt.title(titulo)
    plt.xlabel(columnas_nombre[columna_predictiva])
    plt.ylabel('Precio')
    ax.plot(index, precio)
    ax.plot(index, scikitlearn, color="g")
    ax.plot(index, modelo_manual, color="r")
    ax.legend(['Precio real', 'Scikitlearn', 'Modelo manual'])
    ax.set_ylim(ymin=0)
    plt.show(f)
    
grafica_lineas(precio_real, estimacion_1ra_variable[0], estimacion_1ra_variable[1], columna_1st_predictivo, "Precios para la primera variable")
grafica_lineas(precio_real, estimacion_2nd_variable[0], estimacion_2nd_variable[1], columna_2nd_predictivo, "Precios para la primera variable")
```
